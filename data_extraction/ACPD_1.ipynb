Escriba una función, `csv_to_numpy` que reciba el nombre de un fichero CSV con datos exclusivamente numéricos (enteros) y devuelva un array NumPy estructurado. La primera línea del fichero CSV deberá contener los nombres de columnas. La función además deberá aceptar el parámetro `delimiter`.
</p>
</div>

```python
import numpy as np
import csv

def csv_to_numpy(csv_file, delimiter=","):
    row_matrix = []

    with open(csv_file, "r" ) as file:
        reader = csv.reader(file, delimiter=delimiter)
        for row_num, row in enumerate(reader):
            if row_num == 0:
                header = row
                continue

            row_list = []

            for col_num, col in enumerate(row):
                row_list.append(int(row[col_num]))
            
            row_matrix.append(row_list)
    np_array = np.array(row_matrix)
    return np_array

csv_to_numpy("numbers.csv", delimiter=",")
# array([[5, 0, 6, 5],
#        [7, 4, 6, 0],
#        [7, 3, 6, 7],
#        [4, 7, 3, 7]])
```

```python
import json

def read_jsonl(jsonl):
    dict_list = []

    with open(jsonl, "r") as json_file:
        for line in json_file:
            json_data = json.loads(line)
            dict_list.append(json_data)
            
    return dict_list
    

read_jsonl("books.jsonl")
# [{'title': 'Dune',
#   'author': 'Frank Herbert',
#   'genre': 'Science Fiction',
#   'published_year': 1965},
#   ...
#  {'title': "The Hitchhiker's Guide to the Galaxy",
#   'author': 'Douglas Adams',
#   'genre': 'Science Fiction',
#   'published_year': 1979}]
```

Implemente una función `sevici_disponibles_cerca(fichero_json, latitud, longitud, distancia_km)`, que recibe el fichero json de SEVICI, unas coordenadas, y una distancia máxima en km; y devuelva un listado de las estaciones cercanas que tienen bicicletas disponibles (incluyendo para cada una de ellas: la distancia, el nombre, coordenadas, y número de bicicletas disponibles en este momento). Dichas estaciones deben estar ordenadas por distancia.

Se proporciona a continuación también una función para calcular la distancia, en km, entre dos puntos, dadas su latitud y longitud. Puede usar esta función en su solución.

</p>
</div>

```python
import math

def haversine_dist(lat1, lon1, lat2, lon2):
    """
    Calculates the distance (in km) between two points given their coordinates (latitude and longitude).
    """
    # Earth radius (in km)
    R = 6371.0

    lat1_rad = math.radians(lat1)
    lon1_rad = math.radians(lon1)
    lat2_rad = math.radians(lat2)
    lon2_rad = math.radians(lon2)

    dlat = lat2_rad - lat1_rad
    dlon = lon2_rad - lon1_rad

    # Haversine formula
    a = (math.sin(dlat / 2)**2) + (math.cos(lat1_rad) * math.cos(lat2_rad) * math.sin(dlon / 2)**2)
    
    # c is the angular distance in radians
    c = 2 * math.atan2(math.sqrt(a), math.sqrt(1 - a))

    # Return distance
    return R * c
```

```python
import json

def sevici_disponibles_cerca(file, latitude, longitude, distance_km):
    station_list = []

    with open(file, "r",encoding="utf-8") as json_file:
        data = json.load(json_file)
    data = data["network"]["stations"]

    for station in data:
        station_info = []
        dist = haversine_dist(latitude, longitude, station["latitude"], station["longitude"])

        if dist <= distance_km:
            station_info.append(dist)
            station_info.append(station["name"])
            station_info.append((station["latitude"], station["longitude"]))
            station_info.append(station["free_bikes"])
            station_list.append(station_info)

    return sorted(station_list, key=lambda row: row[0])

sevici_disponibles_cerca('sevici.json', 37.3583071, -5.9875803, 1)
# [
#   [0.10297508977312558, '126_AVENIDA REINA MERCEDES', (37.358414, -5.986423), 13],
#   [0.19958907104307508, '145_AVENIDA REINA MERCEDES', (37.359827, -5.986379), 3],
#  ...
#   [0.9894019243335594, '028_CALLE RAFAEL SALGADO', (37.3602613692166, -5.97665913806671), 17]
# ]
```

Cree una función que reciba el nombre de un equipo, un valor entero $n$ representando el número de goles, y un nombre de fichero. La función debe hacer scraping de la web de Marca (http://www.marca.com/futbol/primera/calendario.html), y almacenar en el fichero en formato CSV la información de aquellos partidos en los que el equipo indicado haya marcado $n$ o más goles. De cada partido debe almacenar cuatro columnas: número de jornada, equipo local, equipo visitante, y resultado.

NOTA: En el notebook utilizado en clase, se veía un ejemplo que puede ser de gran ayuda para resolver este ejercicio.

</p>

</div> 

```python
import csv

def mas_de_n_goles(equipo, n, fichero):
    url = "https://www.marca.com/futbol/primera-division/calendario.html"
    soup = BeautifulSoup(requests.get(url).text)
    jornadas = soup.find_all("div", "jornada")

    jornada_dict ={}

    # Creamos un diccionario con las jornadas y sus partidos
    for jornada in jornadas:
        nombre_jornada = jornada.find("caption").text
        partidos_jornada = jornada.find_all("tr")
        jornada_dict[nombre_jornada] = partidos_jornada

    # Para cada partido en una jornada, obtenemos quien es el local, visitante, y resultado
    
    for nombre_jornada, partidos_jornada in jornada_dict.items():
        results_list = []
        for partido_jornada in partidos_jornada[1:]:
            local = visitante = resultado = ""
            try:
                local = partido_jornada.find("td", "local").text.strip()
                visitante = partido_jornada.find("td", "visitante").text.strip()
                resultado = partido_jornada.find("td", "resultado").text.strip()
                try:
                    goles_local, goles_visitante = resultado.split('-')
                    if (local == equipo and int(goles_local) >= n) or (visitante == equipo and int(goles_visitante) >= n):
                        results_list.append((local, visitante, resultado))
                except:
                    continue 
            except:
                pass
        # Eliminamos la primera posición, que solo tenía el header de la tabla
        jornada_dict[nombre_jornada] = results_list 

    dict_keys = list(jornada_dict.keys())
    for key in dict_keys:
        if len(jornada_dict[key]) == 0:
            del jornada_dict[key]
    
    with open(fichero, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['Jornada', 'Local','Visitante','Resultado'])
        
        for jornada, info in jornada_dict.items():
            jornada_name = jornada
            for i in range(len(info)):
                local, visitante, resultado = info[i]
                writer.writerow([jornada_name, local, visitante, resultado])
            
    return jornada_dict
    
mas_de_n_goles("Sevilla", 3, "sevilla_mas_de_3_goles.csv")
```

```python
%pip install requests
import requests
import pandas as pd

# Si la api key no funciona, obtener una propia en http://www.omdbapi.com/apikey.aspx y reemplzarla. Parece ser que la gratuita solo tiene de vida un dia.

OMDb_API_KEY='f72a310b'

def harry_potter():
    """
        Obtiene las películas de Harry Potter y sus puntuaciones IMDB,
        devolviendo un DataFrame ordenado por puntuación descendente.
    
        Returns:
            pd.DataFrame: DataFrame con columnas 'Movie' y 'Score', ordenado por 'Score' descendente.
    """
    url = f"http://www.omdbapi.com/"
    params = {
        'apikey': OMDb_API_KEY,
        's': 'Harry Potter'
    }
    response = requests.get(url, params=params)
    data = response.json()
    movies = []
    for movie in data['Search']:
        params_movie = {
            'apikey': OMDb_API_KEY,
            'i': movie['imdbID']
        }
        movie_response = requests.get(url, params=params_movie)
        movie_data = movie_response.json()
        movies.append({
            'Movie': movie_data['Title'],
            'Score': float(movie_data['imdbRating'])
        })
    df_movies = pd.DataFrame(movies)
    df_movies.sort_values(by='Score', ascending=False, inplace=True)
    return df_movies

harry_potter()

#     Movie                                               Score
# 0   Harry Potter and the Deathly Hallows: Part 2        8.1
# 8   Harry Potter 20th Anniversary: Return to Hogwarts   8.0
# 9   Harry Potter 20th Anniversary: Return to Hogwarts   8.0
# 2   Harry Potter and the Prisoner of Azkaban            7.9
# 1   Harry Potter and the Sorcerer's Stone               7.7
# 4   Harry Potter and the Goblet of Fire                 7.7
# 7   Harry Potter and the Deathly Hallows: Part 1        7.7
# 6   Harry Potter and the Half-Blood Prince              7.6
# 3   Harry Potter and the Chamber of Secrets             7.5
# 5   Harry Potter and the Order of the Phoenix           7.5
```

Veamos la [API de SpaceX](https://spacex-production.up.railway.app/), en GrapQL, que proporciona información del lanzamiento de naves, incluyendo datos históricos, etc.

Cree una función en Python que reciba un número entero `n`, y devuelva la información de los `n` primeros lanzamientos de SpaceX. Para cada uno de ellos, queremos conocer: nombre de la misión, fecha y hora del lanzamiento (UTC), nombre y tipo de la nave/cohete, y enlace al vídeo de dicho lanzamiento. Dicha información debe devolverse como un DataFrame de $n$ filas y 6 columnas. Para saber dónde encontrar esa información, debe navegar por la documentación de la API, para analizar y conocer qué nos proporciona.

Tenga en cuenta que, en la consulta, el "nodo" `launchesPast` devuelve información sobre lanzamientos en el pasado, mientras que al filtrarlo con `(limit: 10)`, indica que devuelva solo los resultados de los 10 primeros lanzamientos, y no todos.

```python
import requests
import pandas as pd

def ultimos_lanzamientos(n: int):
    """
    Obtiene los últimos n lanzamientos de SpaceX y devuelve un DataFrame con la información relevante.
    Args:
        n (int): Número de lanzamientos a obtener.
    Returns:
        pd.DataFrame: DataFrame con los detalles de los lanzamientos.
    """
    # Endpoint oficial de la comunidad de SpaceX (GraphQL)
    url = 'https://spacex-production.up.railway.app/'
    
    # 'launchesPast' permite obtener datos históricos reales
    # limit: 10 permite acceder a solo 10 registros; modificar como se considere necesario
    query = f"""
    query {{  launchesPast(limit: {n}) {{
        mission_name
        launch_date_utc
        rocket {{
          rocket_name
          rocket_type
        }}
        links {{
          video_link
        }}
    }}
    }}"""
    
    # Hacemos una petición POST
    response = requests.post(url, json={'query': query})

    if response.status_code == 200:
        data = response.json()
        launches = data['data']['launchesPast']
        
        # Procesamos los datos para crear un DataFrame
        records = []
        for launch in launches:
            records.append({
                'Misión': launch['mission_name'],
                'Fecha': launch['launch_date_utc'].split('T')[0],
                'Hora': launch['launch_date_utc'].split('T')[1],
                'Nombre nave': launch['rocket']['rocket_name'],
                'Tipo nave': launch['rocket']['rocket_type'],
                'Video': launch['links']['video_link']
            })
        
        df = pd.DataFrame(records)
        
    else:
        raise ValueError("Error al obtener los lanzamientos")  # Devuelve un DataFrame vacío en caso de error
    return df
    # ...
ultimos_lanzamientos(10)

#     Misión      Fecha       Hora        Nombre nave     Tipo nave   Video
# 0   FalconSat   2006-03-24  22:30:00    Falcon 1        rocket      https://www.youtube.com/watch?v=0a_00nJ_Y88
# 1   DemoSat     2007-03-21  01:10:00    Falcon 1        rocket      https://www.youtube.com/watch?v=Lk4zQ2wP-Nc
# ...
# 8   CRS-1       2012-10-08  00:35:00    Falcon 9        rocket      https://www.youtube.com/watch?v=-Vk3hiV_zXU
# 9   CRS-2       2013-03-01  19:10:00    Falcon 9        rocket      https://www.youtube.com/watch?v=ik0ElKl5kW4
```

Considere la API en https://pokeapi.co/. Implemente una función que, dada una lista con nombres de Pokémons, genere un DataFrame de Pandas con un dataset con información de los mismos. Dicho DataFrame debe contener la siguiente información sobre cada Pokémon: nombre, tipo(s) (como una lista), peso (en kg), altura (en cm), y número de movimientos (*moves*) disponibles. En caso de que alguno(s) de los Pokémon de la lista no existan, simplemente lo obviará, y no lo incluirá en el dataset.

Considere la [documentación](https://pokeapi.co/docs/v2#pokemon) donde se indica los valores devueltos en la respuesta, y las unidades de medida de cada uno de ellos (si aplica).

</p>
</div>

```python
import requests

def info_pokemon(nombre: str):
    """
    Obtiene información de un Pokémon dado su nombre.
    Args:
        nombre (str): Nombre del Pokémon.
    Returns:
        tuple: (nombre_pokemon, tipos, peso, altura, n_moves) o None si no existe.
    """
    url = f"https://pokeapi.co/api/v2/pokemon/{nombre.lower()}"
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        nombre_pokemon = data['name'].capitalize()
        tipos = [tipo['type']['name'] for tipo in data['types']]
        peso = data['weight'] / 10.0  # Convertir a kg
        altura = data['height'] / 10.0  # Convertir a m
        moves = data['moves']
        n_moves = len(moves)
        
        return (nombre_pokemon, tipos, peso, altura, n_moves)
    else:
        return None

# Pruebas
# print(info_pokemon("Pikachu"))
# ('Pikachu', ['electric'], 6.0, 0.4, 105)

# print(info_pokemon("Charizard"))
# ('Charizard', ['fire', 'flying'], 90.5, 1.7, 131)

# print(info_pokemon("Agumon")) # NO es un Pokemon
# None
```

```python
def dataset_pokemon(lista_pokemons: list):
    """
    Crea un DataFrame con información de una lista de Pokémones.
    
    Args:
        lista_pokemons (list): Lista de nombres de Pokémones.
    Returns:
        pd.DataFrame: DataFrame con columnas 'Nombre', 'Tipos', 'Peso (kg)', 'Altura (m)', 'Número de movimientos'. 
    """
    records = []
    for nombre in lista_pokemons:
        info = info_pokemon(nombre)
        if info is not None:
            records.append({
                'Nombre': info[0],
                'Tipos': ', '.join(info[1]),
                'Peso (kg)': info[2],
                'Altura (m)': info[3],
                'Número de movimientos': info[4]
            })
    df = pd.DataFrame(records)
    return df
```
